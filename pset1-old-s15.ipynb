# Problem Quadratic Blues

If we are solving the quadratic equation $x^2 - 2bx + c = 0$, everyone knows the quadratic formula $x_\pm = b \pm \sqrt{b^2 - c}$.  However, consider the accuracy of the $x_-$ root of this equation as $c$ gets smaller:

```julia
b = 1
c = logspace(-20,-1,50) # 50 values from 10^-20 to 10^-1, equally spaced on a log scale
x = b - sqrt(b^2 - c)   # x computed in the default double precision (~15 digits)
set_bigfloat_precision(256) # 256 bits, about 80 decimal digits
xbig = b - sqrt(b^2 - big(c)) # x computed in "bigfloat" precision (~80 digits)
err = float64(abs(x - xbig) ./ abs(xbig)) # relative error, converted to double-precision to plot
using PyPlot
loglog(c, err, "bo-")
xlabel(L"c")
ylabel(L"relative error in $x$")
title(L"standard quadratic formula $x = b - \sqrt{b^2 - c}$")
```

As you can see, once $c$ is comparable to $\epsilon$ for double precision, we have lost *all* our accuracy.  It is easy to see why: $b^2 - c$ loses $c$ completely to roundoff error when $c \ll b^2$, and you eventually get $b^2 \ominus c = b^2$ where $\ominus$ is floating-point subtraction.

There are several ways to fix this.  One possibility is to Taylor-expand the equation in $c$, but this is a bit tricky because you have to be sure to keep enough terms to be accurate to machine precision, and you have to switch back to the regular formula when $c$ is not small (with a carefully selected crossover point).   Instead, we can use this clever rearrangement of the quadratic formula, which is equivalent to dividing the equation by $x^2$ and then applying the quadratic formula to $1/x$:
$$
x_- = \frac{c}{b + \sqrt{b^2 - c}}
$$

```julia
x = c ./ (b + sqrt(b^2 - c))
err = float64(abs(x - xbig) ./ abs(xbig))
loglog(c, err, "bo-")
xlabel(L"c")
ylabel(L"relative error in $x$")
title(L"modified quadratic formula $x = c / (b + \sqrt{b^2 - c})$")
```

We now achieve nearly machine precision or better across the whole range of $c$!

# Newtonish

Suppose we are trying to find a root $f(x)=0$ and we are given ways to compute $f$, $f'$, and $f''$.   We approximate $f(x)$ by the first two terms in the Taylor series:
$$
f(x - \delta) \approx f(x) - \delta f'(x) + \frac{\delta^2}{2} f''(x) = 0
$$
and set the right-hand side to zero to find an approximate root.  Solving the quadratic equation and eliding the $(x)$ arguments, we obtain an approximate step to the root:
$$
\delta = \frac{f' \pm \sqrt{f'^2 - 2ff''}}{f''}
$$
Close to the root, $f$ is small and therefore the discriminant will be positive and we will get real roots.   We want the minus root of this quadratic, because we want the *nearest* root (the smallest $\delta$).  But this will run into exactly the roundoff problems noted above, so we rewrite the desired solution as:
$$
\delta = \frac{2f}{f' + \sqrt{f'^2 - 2ff''}}
$$
If we are not close to the root, so that the discriminant is negative, we will just take an ordinary Newton step $\delta = f/f'$.  Notice that this is also approximately the step we take when $f$ is very small.

```julia
# take n Newtonish steps starting at x, given functions f(x),f′(x),f″(x) 
function newtonish(f,f′,f″, x, n)
    for i = 1:n
        fx = f(x)
        f′x = f′(x)
        f″x = f″(x)
        discrim = f′x^2 - 2*fx*f″x
        δ = discrim < 0 ? fx/f′x : 2fx / (f′x + sqrt(discrim))
        x = x - δ
    end
    return x
end
```

Now, let's try it out on the cube root function, i.e. to find the root $1$ of $f(x) = x^3 - 1$.

```julia
using Interact
set_bigfloat_precision(20000)
@manipulate for n = 0:9
    # output the number of accurate decimal digits, via log10 of the error:
    -float64(log10(abs(newtonish(x->x^3-1, x->3x^2, x->6x, big(2), n) - 1)))
end
```

It is easy to see that the number of digits is roughly tripling on each iteration: **cubic convergence**!   It is straightforward to prove this (once you are sufficiently close to the root).

In practice, the quadratic convergence of the ordinary Newton's method is so fast that you hardly ever resort to higher-order methods.  The hard part is finding a starting point sufficiently close to the root that you want, not converging fast once you get there.

Also, in high-dimensional root-finding problems, $f'$ is replaced by the Jacobian matrix, while $f''$ is a rank-3 tensor ("3d matrix") that would be expensive to compute and work with.
