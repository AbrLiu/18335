## Problem 1(a): QR, RQ

Consider the following iteration, repeatedly factorizing $A=QR$ and then forming $A \leftarrow RQ$:

```julia
X = rand(5,5)
A = X + X'
display(A)
for n = 0:10
    Q,R = qr(A)
    A = R*Q
    display(A)
end
```

If we compare the results above with the eigenvalues of the starting matrix $X+X^*$, we can start to see that it is converging to a diagonal matrix of the eigenvalues, sorted in descending order by absolute value:

```julia
eigvals(X + X')
```

However, it is converging too slowly to see this with only 10 iterations.  Let's run it for 1000 iterations instead, and then the eigenvalues and the diagonal nature of the matrix are obvious:

```julia
A = X + X'
for n = 0:1000
    Q,R = qr(A)
    A = R*Q
end
A
```

We can also plot the convergence rate, for fun:

```julia
using PyPlot

A = X + X'
ns = 0:100
Adiags = zeros(length(ns),size(A,1))
for n in ns
    Adiags[n+1,:] = diag(A)
    Q,R = qr(A)
    A = R*Q
end
semilogy(ns, abs(Adiags .- sort(eigvals(X + X'), by=abs, rev=true)'), ".-")
legend(["\$|a_{$n$n} - \\lambda_$n|\$" for n in 1:size(A,1)])
xlabel("QR iterations")
```

Notice that the first eigenvalues to converge are $\lambda_1$ and $\lambda_5$, the *biggest and smallest* magnitudes.  This is because, as explained in class, QR is equivalent to both simultaneous power iteration and simultaneous inverse iteration, which converge the extremal eigenvalues first.

## Problem 3: Matrix addition and the CPU

Here we compute matrix addition $C = A + B$ for $m \times n$ matrices, which is trivial to implement: just two loops over $i$ and $j$ to compute $c_{ij} = a_{ij} + b_{ij}$ for each entry.  There is no possibility of temporal locality in this problem, because each matrix element is used exactly once.  However, the ordering of the operations still matters because of *spatial locality* (primarily because of *cache lines*).   In particular, we will look at the influence of the loop order: does it matter (for performance) if we perform the $i$ loop first (outermost), then the $j$ loop, or vice versa?

### Two matrix-addition implementations

Here are two simple matrix-addition implementations in Julia that differ only in their loop ordering.  In both cases we have used the `@inbounds` macro to avoid the overhead of bounds-checking array accesses (which makes a slight performance difference in inner loops), and we implement "mutating" variants that require the output array `C` to be pre-allocated (to avoid the overhead of array allocation during benchmarking).

```julia
function matadd1!(C, A, B)
    m,n = size(A)
    (m,n) == size(B) || error("matrix sizes must match")
    for i = 1:m
        for j = 1:n
            @inbounds C[i,j] = A[i,j] + B[i,j]
        end
    end
    return C
end
matadd1(A,B) = matadd1!(similar(A, promote_type(eltype(A),eltype(B))), A,B)

# same as matadd1 except the loop order is reversed
function matadd2!(C, A, B)
    m,n = size(A)
    (m,n) == size(B) || error("matrix sizes must match")
    for j = 1:n
        for i = 1:m
            @inbounds C[i,j] = A[i,j] + B[i,j]
        end
    end
    return C
end
matadd2(A,B) = matadd2!(similar(A, promote_type(eltype(A),eltype(B))), A,B)
```

```julia
# correctness check:
A = rand(5,7)
B = rand(5,7)
norm(matadd1(A,B) - (A+B)), norm(matadd2(A,B) - (A+B))
```

You can also use the `?` command in Julia to look at how the built-in `A+B` is implemented, since `?A+B` will give a link to the source code of the `+` function that is used for these argument types:

```julia
?A+B
```

### part (a): Benchmarking matrix addition

Now, let's benchmark `matadd1!` and `matadd2!`, along with the built-in `A+B` (which allocates a new array as output each time you call it), for a variety of $n\times n$ matrix sizes:

```julia
N = round(Int, linspace(10, 1000, 200))
t0 = Float64[]
t1 = Float64[]
t2 = Float64[]
for n in N
    A = rand(n,n)
    B = rand(n,n)
    C = similar(A)
    push!(t0, @elapsed A+B) # note: does not preallocate output
    push!(t1, @elapsed matadd1!(C,A,B))
    push!(t2, @elapsed matadd2!(C,A,B))
end
```

and plot the results as gigaflops ($n^2$ divided by time in nanoseconds) vs. $n$:

```julia
using PyPlot

plot(N, N.^2 ./ t0 * 1e-9, "k.-")
plot(N, N.^2 ./ t1 * 1e-9, "bo-")
plot(N, N.^2 ./ t2 * 1e-9, "rs-")
xlabel(L"n")
ylabel(L"gflops ($n^2/t$)")
legend(["A+B","matadd1!","matadd2!"])
```

What's going on here is that [Julia arrays are stored in column-major order](http://docs.julialang.org/en/latest/manual/performance-tips/#access-arrays-in-memory-order-along-columns), i.e. with the columns contiguous in memory.

Hence, you will obtain greater *spatial* locality (better cache-line usage) by `matadd2`, which completely adds one pair of columns before moving to the next one.  In contrast, `matadd1` adds by rows, which accesses the memory in discontiguous fashion and does not exploit the cache line.

### part (b): preallocation

Notice that the built-in `A+B` is actually *slower* than `matadd2`.  `A+B` is accessing the arrays in the correct (contiguous) order, but it allocates a new matrix every time you call it.  The penalty for that, rather than working in-place on a preallocated array `C`, is significant for an operation as cheap as `A+B`.

Let's try the benchmark again, but with all three lacking pre-allocated output:

```julia
N = round(Int, linspace(10, 1000, 200))
t0 = Float64[]
t1 = Float64[]
t2 = Float64[]
for n in N
    A = rand(n,n)
    B = rand(n,n)
    push!(t0, @elapsed A+B) # note: does not preallocate output
    push!(t1, @elapsed matadd1(A,B))
    push!(t2, @elapsed matadd2(A,B))
end

plot(N, N.^2 ./ t0 * 1e-9, "k.-")
plot(N, N.^2 ./ t1 * 1e-9, "bo-")
plot(N, N.^2 ./ t2 * 1e-9, "rs-")
xlabel(L"n")
ylabel(L"gflops ($n^2/t$)")
legend(["A+B","matadd1","matadd2"])
title("benchmarks without preallocation")
```

As expected, the built-in `A+B` is now comparable to `matadd2`.

### part (c): putting it all together

If we compute `C=A+3B+4A.^2` by that exact expression, Julia (or Matlab or Numpy or similar languages) will break it up into a sequence of steps, roughly:

 1. allocate a temporary matrix `T1` and assign it to the result of `A.^2`
 2. allocate/compute `T2 = 4T1`
 3. allocate/compute `T3 = 3B`
 4. allocate/compute `T4 = A+T3`
 5. allocate/compute `C = T4+T2`

In particular, note that each individual operation will generally occur in its *own loop*, allocating its *own temporary* array.  That is because `A+3B+4A.^2` is not a built-in function in Julia or any other language — only operations like `+` and `*` and `.^` are built-in, so it has to compute these one at a time and then combine the results.  This means that there is a lot of overhead and wasted memory from the allocation, and poor spatial/temporal locality from the repeated passes over different arrays.

In principle, there is a compiler optimization called [loop fusion](http://en.wikipedia.org/wiki/Loop_fusion) that can combine these operations, and similar optimizations to eliminate the need for temporary arrays.  But they are often not available for sequences of operations like the ones above unless array operations are built into the *language* (so that the *compiler* knows what they mean, e.g. as in Fortran 90), as opposed to just being functions of a language *library* (e.g. the Julia built-in library or the NumPy library in Python) of a general-purpose language.

You can therefore, do much better by just writing your own loops — provided that you have a language like Julia where loops are fast.  (In Matlab or Python, you would have to write your own loops in C or Cython.)  Of course, this takes (a little) more code, and is more verbose, so you shouldn't bother unless performance really matters (google "premature optimization").  Let's try it to see how much of a difference it makes:

```julia
function myfunc!(C, A, B)
    length(A) == length(B) == length(C) || error("array length mismatch")
    for i = 1:length(C) # if we loop over a single index, Julia will access the data contiguously
        @inbounds a = A[i]
        @inbounds C[i] = a + 3B[i] + 4a^2
    end
    return C
end
myfunc(A, B) = myfunc!(similar(A, promote_type(eltype(A),eltype(B))), A,B)
```

```julia
# correctness check:
A = rand(5,7)
B = rand(5,7)
norm(myfunc(A,B) - (A+3B+4A.^2))
```

```julia
N = round(Int, linspace(10, 1000, 200))
t0 = Float64[]
t1 = Float64[]
t2 = Float64[]
for n in N
    A = rand(n,n)
    B = rand(n,n)
    C = similar(A)
    push!(t0, @elapsed A+3B+4A.^2)
    push!(t1, @elapsed myfunc(A,B))
    push!(t2, @elapsed myfunc!(C,A,B))
end

plot(N, 5N.^2 ./ t0 * 1e-9, "k.-")
plot(N, 5N.^2 ./ t1 * 1e-9, "bo-")
plot(N, 5N.^2 ./ t2 * 1e-9, "rs-")
xlabel(L"n")
ylabel(L"gflops ($5n^2/t$)")
legend(["A+3B+4A.^2","myfunc","myfunc!"])
title("benchmarking A+3B+4A.^2 variants")
```

Because `myfunc` is more expensive than simple addition, the overhead of allocation is less important, so we don't see as much benefit from the in-place `myfunc!` variant (although its performance is more consistent: memory allocation adds some "randomness" into the timing).   Both `myfunc` variants are so much faster than `A+3B+4A.^2` that we can barely see the latter!  Let's plot the ratio instead:

```julia
plot(N, t0 ./ t2, "ro-")
xlabel(L"n")
ylabel("speedup of myfunc! / A+3B+4A.^2")
grid()
```

Typically a 10–15× speedup!

As a general rule, with one needs to **unlearn the instinct** instilled by languages such as Matlab or Python or R that *efficiency = clever use of library functions*, which turns all optimization questions into *is there a built-in function for X* (and if the answer is "no" then you are out of luck unless you are willing to write C).   Loops are fast, and you can easily beat general-purpose library functions with your own special-purpose code.

(Exploiting highly optimized, robust, well-tested numerical libraries is still a good idea in *any* language.  But, when speed is critical, it is useful to remember that a *specialized* function can almost always beat a *general-purpose* function or a sequence thereof.)
